```python
import random
import csv
import re

l = ['id','text','label']
tmp =[]
with open('./datasets.tsv', 'w') as f:
    writer = csv.writer(f,delimiter="\t")
    writer.writerow(l)
with open('./text/livedoor.tsv') as f1:
    l=f1.readlines()
    leng = len(l)
    L=random.sample(l,leng)
    for row in L:
        row_l = row.split('\t')
        tmp.insert(0,re.sub("\\D", "", row_l[0]))
        tmp.append(re.sub('\u3000', "", row_l[1].split(' ')[0]))
        tmp.append(re.sub('\n', "", row_l[-1]))
        with open('./datasets.tsv', 'a') as f2:
            writer = csv.writer(f2,delimiter="\t")
            writer.writerow(tmp)
        tmp =[]
```

```python
l = ['id','text','label']

with open('./text/dataset/train.tsv', 'w') as f:
    writer = csv.writer(f,delimiter="\t")
    writer.writerow(l)
with open('./datasets.tsv') as f1:
    l2=f1.readlines()
    leng = len(l2)
    t_l = round(leng*4/5)
    L=l2[0:t_l]
    with open('./text/dataset/train.tsv', 'a') as f2:
        writer = csv.writer(f2)
        writer.writerow(L)
         
with open('./text/dataset/dev.tsv', 'w') as f:
    writer = csv.writer(f,delimiter="\t")
    writer.writerow(l)        
with open('./datasets.tsv') as f1:
    l2=f1.readlines()
    t_d = round(leng*1/10)
    L=l2[t_l+1:t_l+t_d]
    with open('./text/dataset/dev.tsv', 'a') as f2:
        writer = csv.writer(f2)
        writer.writerow(L)
           
with open('./text/dataset/test.tsv', 'w') as f:
    writer = csv.writer(f,delimiter="\t")
    writer.writerow(l)        
with open('./datasets.tsv') as f1:
    l2=f1.readlines()    
    t_d = round(leng*1/10)
    L=l2[t_l+t_d+1:leng]
    with open('./text/dataset/test.tsv', 'a') as f2:
        writer = csv.writer(f2)
        writer.writerow(L)
```
